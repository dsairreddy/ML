```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/DATA.csv')
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


X = df[['Volume', 'Weight']]
Y = df['CO2']
```

```python


x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(y_pred)
y_test
```

```python
def co2Emission():
    LISTofMODELS = df[['Car','Model']].values.tolist()

    print("Available car models:")
    for i, x in enumerate (LISTofMODELS):
        print(f'{i+1:}.{x[0]} - {x[1]}')
    SelectModelIndex = int(input('Enter the Car model index: '))-1
    SelectModel = LISTofMODELS[SelectModelIndex][1]
    while True:
        try:
            co2_emission = df.loc[df['Model'] == SelectModel, 'CO2'].values[0]
            return co2_emission
        except IndexError:
            return "Model not found in the dataset."
        break # This break statement was inside the while loop, but outside the try/except block.
co2_emission_value = co2Emission()


print(f"CO2 emission for selected model: {co2_emission_value}")
```
