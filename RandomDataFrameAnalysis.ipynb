```python

import random
import pandas as pd

a1 = []
a2 = []
a3 = []
a4 = []
a5 = []
for _ in range(1000):
    a1.append(int(random.randint(0, 100)))
    a2.append(int(random.randint(50, 70)))
    a3.append(int(random.randint(10, 20)))
    a4.append(int(random.randint(5, 10)))
    a5.append(int(random.randint(20, 100)))
```

```python
df = {
    'grade': a1,
    'weight': a2,
    'friends': a3,
    'cousins': a4,
    'projects': a5
}
df1=pd.DataFrame(df)
```
